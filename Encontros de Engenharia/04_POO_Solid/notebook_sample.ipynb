```python
import pandas as pd
import numpy as np
```

```python
exchange_rates_df = pd.read_csv(
    filepath_or_buffer="files/exchange_rates.csv",
    sep=",",
    dtype={
        "Country/Currency": str,
        "currency": str,
        "value": np.float64
    },
    parse_dates=["date"]
)
```

```python
exchange_rates_df.rename(
    columns={
        "Country/Currency": "nm_country_currency",
        "currency": "sg_currency",
        "value": "nb_value",
        "date": "dt_extraction"
    }, 
    inplace= True
)
```

```python
exchange_rates_df
```

```python
sample_currency_data = pd.read_csv(
    filepath_or_buffer="files/SampleCurrencyData.txt",
    sep="\t",
    header=0,
    names=["AverageRate", "CurrencyID", "CurrencyDate", "EndOfDayRate"],
    dtype={
        "AverageRate": np.float64,
        "CurrencyID": str,
        "EndOfDayRate": np.float64
    },
    parse_dates=["CurrencyDate"]
)
```

```python
sample_currency_data.rename(
    columns={
        "AverageRate": "nb_average_rate",
        "CurrencyID": "sg_currency",
        "EndOfDayRate": "nb_end_day_rate",
        "CurrencyDate": "dt_currency_date"
    }, 
    inplace= True
)
```

```python
sample_currency_data
```

```python
import requests

api_key = ""
response = requests.get(f"https://api.freecurrencyapi.com/v1/latest?apikey={api_key}")
response.json()
```

```python
df_currency_converter = pd.DataFrame(response.json()).reset_index()
```

```python
df_currency_converter.rename(columns={
        "index": "nm_currency",
        "data": "nb_value"
    },
    inplace=True
)
```

```python
from datetime import datetime 
df_currency_converter["dt_extraction"] = datetime.now()
```
